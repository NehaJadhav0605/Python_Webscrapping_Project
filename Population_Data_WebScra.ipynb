```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.worldometers.info/world-population/"
response=requests.get(url)
response
```

```python
response.text
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup.prettify()
```

```python
with open("File.html", "w") as f:
  f.write(response.text)
```

```python
title = soup.find("h2", {"id":"table-historical"})

if title:
    population = title.text.strip()
    print(population)
else:
    print("Title not found")
```

```python
title.text
```

```python
population_data=soup.find("table")
population_data
```

```python
data=population_data.find("tbody")
data
```

```python
rows=data.find_all("td")
rows
```

```python
info = []

for i in rows:
    info.append(i.text.strip())

print(info)
```

```python
row_wise_data = []
row_length = 7

for i in range(0, len(info), row_length):
    row_wise_data.append(info[i:i+row_length])
```

```python
Year = []
Population = []
Yearly_perc_change = []
Yearly_change = []
Median_age = []
Fertility = []
Density = []

for i in row_wise_data:
    Year.append(i[0])
    Population.append(i[1])
    Yearly_perc_change.append(i[2])
    Yearly_change.append(i[3])
    Median_age.append(i[4])
    Fertility.append(i[5])
    Density.append(i[6])

print(Year)
print(Population)
print(Yearly_perc_change)
print(Yearly_change)
print(Median_age)
print(Fertility)
print(Density)
```

```python
p_data={
    "Year": Year,
    "Population": Population,
    "Yearly_%_change": Yearly_perc_change,
    "Yearly_change": Yearly_change,
    "Median_age": Median_age,
    "Fertility": Fertility,
    "Density": Density
}
```

```python
print(p_data)
```
